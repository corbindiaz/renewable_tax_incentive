```python
# Load the dataset
nrel_laws_incentives = pd.read_csv("nrel_laws_incentives.csv")

# Define state abbreviation mapping
state_mapping = {
    'OH': 'Ohio',
    'OK': 'Oklahoma',
    'OR': 'Oregon',
    'PA': 'Pennsylvania',
    'RI': 'Rhode Island',
    'SD': 'South Dakota',
    'TN': 'Tennessee'
}

# Step 1: Extract the year from the 'Enacted Date' column
nrel_laws_incentives['Year'] = pd.to_datetime(nrel_laws_incentives['Enacted Date'], errors='coerce').dt.year

# Step 2: Filter for the specified states (using the state abbreviations in the 'State' column)
filtered_df = nrel_laws_incentives[nrel_laws_incentives['State'].isin(state_mapping.keys())].copy()  # Make a copy

# Step 3: Rename the states to their full names
filtered_df.loc[:, 'State'] = filtered_df['State'].map(state_mapping)

# Step 4: Combine the metadata columns into a single 'Metadata' column
filtered_df.loc[:, 'Metadata'] = filtered_df[['Title', 'Text', 'Type', 'Agency', 'Sequence Number', 
                                               'Technology Categories', 'Incentive Categories', 
                                               'Regulation Categories', 'User Categories', 
                                               'References', 'Status']].apply(
    lambda row: ', '.join(row.dropna().astype(str)), axis=1)

# Step 5: Convert the 'Year' column to integers (remove decimals)
filtered_df.loc[:, 'Year'] = filtered_df['Year'].astype('Int64')  # 'Int64' type to handle NaNs as well

# Step 6: Select only the columns 'Year', 'State', and 'Metadata'
final_df = filtered_df[['Year', 'State', 'Metadata']]

# Display the final DataFrame
print(final_df.head())
```
